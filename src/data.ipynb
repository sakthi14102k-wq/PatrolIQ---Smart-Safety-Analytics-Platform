```python
import pandas as pd
from urllib.parse import quote

def download_month_stratified_sample(
        start_year=2001,
        end_year=2025,
        rows_per_month=2400, 
        output_file="chicago_crime_stratified_months.csv"
):
    base_url = "https://data.cityofchicago.org/resource/ijzp-q8t2.csv"
    all_data = []

    for year in range(start_year, end_year + 1):
        for month in range(1, 13):
            # Format month to have leading zero (e.g., '01', '02')
            month_str = f"{month:02d}"
            
            # Determine the last day of the month (simplified)
            last_day = "31" if month in [1, 3, 5, 7, 8, 10, 12] else "30"
            if month == 2: last_day = "28" # Simple leap year ignore for sampling
            
            print(f"Downloading: {year}-{month_str}")

            # Socrata Query (SoQL) for the specific month
            where_query = f"date between '{year}-{month_str}-01T00:00:00' and '{year}-{month_str}-{last_day}T23:59:59'"
            encoded_query = quote(where_query)

            # Construct URL with limit per month
            url = f"{base_url}?$where={encoded_query}&$limit={rows_per_month}"

            try:
                df_month = pd.read_csv(url)
                if not df_month.empty:
                    all_data.append(df_month)
            except Exception as e:
                print(f"Error downloading {year}-{month}: {e}")

    print("Combining monthly datasets...")
    final_df = pd.concat(all_data, ignore_index=True)

    # Shuffle to mix months and years
    final_df = final_df.sample(frac=1).reset_index(drop=True)
    final_df.to_csv(output_file, index=False)

    print(f"\nSUCCESS. Saved {len(final_df)} rows to {output_file}")
    return final_df

# Run the function
df = download_month_stratified_sample(rows_per_month=1700)
```
